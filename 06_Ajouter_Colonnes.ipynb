```python
import pandas as pd 
df = pd.read_csv('data.csv')
df
```

```python
df["tax"].fillna(0,inplace = True)
```

```python
df
```

```python
df.price_paid = df.price_paid.apply(lambda x : x.replace("$",""))
df.price_paid = df.price_paid.astype(float)
df
```

```python
df["price_total"] = df["price_paid"] * (1-df["tax"]/100)
df
```

```python
df["price_total"] = df["price_paid"] - (df["price_paid"]*df["tax"]/100)
```
